```python
import pandas as pd
df=pd.read_csv("train.csv")
df.head()
```

```python

a={'male':1,'female':2}
df.Sex=[a[item] for item in df.Sex]
df.sample()
```

```python
df.drop(['Name','Ticket','Fare','Cabin','Age','Embarked'],axis=1)
```

```python
X=df[['PassengerId','Pclass','Sex','SibSp','Parch']]
Y=df['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X,Y)
```

```python
model.score(X,Y)
```
